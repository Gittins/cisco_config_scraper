Use list comprehension to iterate through each list, comparing each item in objct against every object in conf. The ".endswith()" method tells it to only match a list item that is ends with exactly the list value of objct.

This is required as we were seeing DM_INLINE_NETWORK_2 matching DM_INLINE_NETWORK_2, DM_INLINE_NETWORK_20, DM_INLINE_NETWORK_21, DM_INLINE_NETWORK_200, DM_INLINE_NETWORK_202, DM_INLINE_NETWORK_222, etc.
